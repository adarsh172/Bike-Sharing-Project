## Step 3: Data Preparation

### Dummy Variable Creation

```python
season = pd.get_dummies(df['season'], drop_first = True)
weather = pd.get_dummies(df['weathersit'], drop_first = True)
month = pd.get_dummies(df['mnth'], drop_first = True)
weekday = pd.get_dummies(df['weekday'], drop_first = True)
```

```python
# Creating a new dataframe called bike where season, month, weather and weekday dataframe is being added to df.
bike = pd.concat([df,season,month,weather,weekday], axis = 1)
bike.head()
```

```python
#deleting the unnecessry column season, mnth, weathersit and weekday as the respective values are already populated as binary columns data
bike.drop(['season','mnth','weathersit','weekday'],axis=1,inplace=True)
bike.shape
```

## Step 4: Model Building

```python
#Machine learning Libraries
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt
```

```python
# We specify this so that the train and test data set always have the same rows, respectively
# Based on the 70% - 30% split between train and test dataset we have 510 rows in train dataset and 220 in test dataset
np.random.seed(0)
bike_train, bike_test = train_test_split(bike, train_size = 0.7, random_state = 300)
```

```python
bike_train.shape
```

```python
bike_test.shape
```

```python
# Rescaling using MinMaxCcaler
scaler = MinMaxScaler()
num_vars = ['temp','hum','windspeed','cnt']
bike_train[num_vars] = scaler.fit_transform(bike_train[num_vars])
bike_train.head()
```

```python
y_train = bike_train.pop('cnt')
X_train = bike_train
```

## RFE
#### Recursive feature elimination: We will be using the LinearRegression function from SciKit Learn for its compatibility with RFE

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
# Init the transformer
rfe = RFE(estimator=LinearRegression(), n_features_to_select=15)

# Fit to the training data
rfe = rfe.fit(X_train, y_train)
```

```python
# Checking which parameters have been selected in that list of 15
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# storing the selected 15 variables in col list
col = X_train.columns[rfe.support_]
col
```

```python
# checking which columns have been eleminated 
X_train.columns[~rfe.support_]
```

```python
# Creating X_train dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Function for VIF Calculation

def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# the feature variables and their respective VIFs
calculateVIF(X_train_rfe)
```

### Humidity can be removed from train set as it has high VIF value

```python
X_train_new = X_train_rfe.drop(['hum'], axis = 1)
# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

### Now almost all the variables are well and good according to there Vif Scores we wont remove anymore variables

```python
# Add a constant
X_train_lm = sm.add_constant(X_train_new)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()

# Print a summary of the linear regression model obtained
print(lr.summary())
```

## INSIGHTS:
#### 1.) R-squared value is really good (85.6%) it means the model is explaining 85.6% of variance in data 
#### 2.) Adjusted  R squared value is also good (85.2%) it means that our model has perfect no. of variables and there is no penalisation due to excess variables.
#### 3.)prob(F-statistic) is very very low it means that our model is fit very well and it is not by chance.
#### 4.)P values of the variables taken in train data are well under 0.05 hence we can say that all  the varaibles are significant.

## Step 5: Model Validation

```python
sm.graphics.plot_ccpr(lr, 'temp')
plt.show()
```

```python
sm.graphics.plot_ccpr(lr, 'windspeed')
plt.show()
```

### 1.)The above plots represents the linear relationship between the model and the predictor variables.

```python
y_train_pred = lr.predict(X_train_lm)
residual = y_train - y_train_pred
sns.scatterplot(y_train,residual)
plt.plot(y_train,(y_train - y_train), '-r')
plt.xlabel('Count')
plt.ylabel('Residual')
plt.show()
```

### 2.)  There is no visible pattern in residual values, thus homoscedasticity is well preserved

```python
print('The Durbin-Watson value for Final Model lr is',round(sm.stats.stattools.durbin_watson((y_train - y_train_pred)),4))
```

### 3.)Autocorrelation refers to the fact that observations errors are correlated. To verify that the observations are not auto-correlated, I used the Durbin-Watson test. The test will output values between 0 and 4. The closer it is to 2, the less auto-correlation there is between the various variables.
### There is almost no autocorrelation.

```python
res = y_train-y_train_pred

# Plotting the histogram of the error terms
fig = plt.figure()
sns.distplot((res), bins = 20)
fig.suptitle('Error Terms')                  
plt.xlabel('Errors')                         
plt.show()
```

### 4.)Based on the histogram, we can conclude that error terms are following a normal distribution with mean at zero.

## Step 6: Making Predictions using final model
### Scaling bike_test dataframe

```python
num_vars = ['temp', 'hum', 'windspeed','cnt']
bike_test[num_vars] = scaler.transform(bike_test[num_vars])
bike_test.head()
```

```python
y_test = bike_test.pop('cnt')
X_test = bike_test
```

```python
#Selecting the variables that were part of final model.
col1=X_train_new.columns

X_test=X_test[col1]

# Adding constant variable to test dataframe
X_test_lm = sm.add_constant(X_test)

X_test_lm.info()
```

```python
# Making predictions using the final model (lr)

y_pred = lr.predict(X_test_lm)
```

## Step 7: Model Evaluation

```python
# plotting y_test and y_pred to understand the spread
fig = plt.figure()
plt.scatter(y_test, y_pred,alpha = 0.5, color='g')
fig.suptitle('y_test vs y_pred')             
plt.xlabel('y_test')                          
plt.ylabel('y_pred')
```

```python
r2 = round(r2_score(y_test, y_pred),4)
print("R-Squared:",r2)
# n is number of rows in test dataset
n = X_test.shape[0]

# Number of features (predictors, p) is the shape along axis 1
p = X_test.shape[1]

# We find the Adjusted R-squared using the formula
adjusted_r2 = round(1-(1-r2)*(n-1)/(n-p-1),4)
print("Adjusted R-Squared:",adjusted_r2)
RMSE = round(sqrt(mean_squared_error(y_test, y_pred)),4)
print("RMSE:",RMSE)
MAE = round(mean_absolute_error(y_test, y_pred),4)
print("MAE:",MAE)
```

## INSIGHTS:
#### 1.)  R-squared and adjusted R-squared for train data were 85.6% and 85.2% respectively and those for test data are 78.9% and 77.4% So, there is not a great difference and the model seems to be predict good on test data.
#### 2.) The Root Mean Squared Error value for the test dataset based on final model is 0.1022 and Mean Absolute Error is 0.0738, which indicates that the model is really good
